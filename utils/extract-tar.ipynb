```python
import os
import tarfile

base = 'drive/MyDrive/Colab Notebooks/NFFA-SEM/'
# Set your paths
original_dir = base + 'data/original'
processed_dir = base + 'data/processed'

# Make sure 'processed' directory exists
os.makedirs(processed_dir, exist_ok=True)

# Loop through all .tar files in 'original'
for filename in os.listdir(original_dir):
    if filename.endswith('.tar'):
        tar_path = os.path.join(original_dir, filename)
        target_folder = os.path.join(processed_dir, filename.replace('.tar', ''))

        # Make a folder for each extracted dataset
        os.makedirs(target_folder, exist_ok=True)

        # Extract
        with tarfile.open(tar_path) as tar:
            tar.extractall(path=target_folder)

        print(f'Extracted {filename} to {target_folder}')
```

```python
import os
import shutil

# ベースディレクトリ
base_dir = '/content/drive/MyDrive/Colab Notebooks/NFFA-SEM/data/processed'

# 全てのサブフォルダをチェック
for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)

    # フォルダである場合だけ処理
    if os.path.isdir(folder_path):
        nested_path = os.path.join(folder_path, folder)

        # 中に同名のフォルダがあるかチェック
        if os.path.isdir(nested_path):
            print(f"Fixing: {folder}")

            # 中身を親に移動
            for item in os.listdir(nested_path):
                src = os.path.join(nested_path, item)
                dst = os.path.join(folder_path, item)
                shutil.move(src, dst)

            # 元の空フォルダを削除
            os.rmdir(nested_path)

print("✅ All folders are fixed!")
```
